# TO SCRAPE EACH RELEVANT URL AND TO APPEND IT INTO A DATAFRAME (OUTDATED)

```python
date_list = []
date_list2 = []
date_list3 = []
countries = []
total_cases = []
no_deaths = []
no_recovered = []
col_5 = []

for url in tag_list[0:20]:
    link = 'https://www.who.int' + url
    html = urlopen(link)
    soup = BeautifulSoup(html, 'lxml')
    table = soup.find('div', id='primary')
    
    dates = url.split('/')
    date = dates[4]
    
    date_list.append(date)
    rows = table.find_all('tr') 
    
    lst1 = []
    lst2 = []
    lst3 = []
    lst4 = []
    lst5 = []

    for row in rows:

        try:
            lst1.append(row.find_all('td')[0].text)
            countries.append(lst1)
        except:
            continue

        try:
            lst2.append(int(row.find_all('td')[1].text))
            total_cases.append(lst2)
        except:
            continue

        try:
            lst3.append(int(row.find_all('td')[2].text))
            no_deaths.append(lst3)
        except:
            continue

        try:
            lst4.append(int(row.find_all('td')[3].text))
            no_recovered.append(lst4)
        except:
            continue

        try:
            lst5.append(row.find_all('td')[4].text)
            col_5.append(lst5)
        except:
            continue

#     cols = {"Countries": lst1, 'Total Cases': lst2, 'Deaths':lst3}
#     df = pd.DataFrame(cols)
    print('list1 :'+str(len(lst1)))
    print('list2 :'+str(len(lst2)))

    print(lst1)
    print(lst2)




```

# DATAFRAME FOR 8TH MARCH TO 9TH APR

```python
df = pd.DataFrame()
for url in tag_list[0:21]:
    
    link = 'https://www.who.int' + url
    html = urlopen(link)
    soup = BeautifulSoup(html, 'lxml')
    table = soup.find('table')
    rows = table.find_all('tr')
    dates = url.split('/')
    date = dates[4]
    for row in rows:
        cols = row.find_all('td')
        cols=[x.text.strip() for x in cols]
        #print(cols)

        df = df.append({"Date":date,"Country": cols[0],"Cumulative Cases":cols[1], "No_Deaths":cols[2],"Local Trans":cols[3]}, ignore_index=True)

df
        
    
```

```python
df2 = pd.DataFrame()
for url in tag_list[21:33]:

    link = 'https://www.who.int' + url
    html = urlopen(link)
    soup = BeautifulSoup(html, 'lxml')
    table = soup.find('table')
    rows = table.find_all('tr')
    dates = url.split('/')
    date = dates[4]
    for row in rows:
        cols = row.find_all('td')
        cols=[x.text.strip() for x in cols]
        
        if len(cols)<6:
            continue
        else:
            df2 = df2.append({"Date":date,"Country": cols[0],"Cumulative Cases":cols[1], "No_Deaths":cols[3],"No Recovered":cols[4],"Local Trans":cols[5]}, ignore_index=True)

df2
```

```python
df3 = pd.DataFrame()
for url in tag_list[33:]:
    
    link = 'https://www.who.int' + url
    html = urlopen(link)
    soup = BeautifulSoup(html, 'lxml')
    table = soup.find('table')
    rows = table.find_all('tr')
    dates = url.split('/')
    date = dates[4]
    for row in rows:
        cols = row.find_all('td')
        cols=[x.text.strip() for x in cols]
        
        if len(cols)<6:
            continue
        else:
            df3 = df3.append({"Date":date,"Country": cols[0],"Cumulative Cases":cols[1], "No_Deaths":cols[3],"No Recovered":cols[4],"Local Trans":cols[5]}, ignore_index=True)

df3
```

# Appending all the dataframes together

```python
df_all = df.append(df2,ignore_index=True)
df_all = df.append(df3, ignore_index=True)
df_all
df_all.to_csv('SARS_data.csv')
```
